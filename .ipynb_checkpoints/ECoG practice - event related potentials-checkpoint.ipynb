```python
import numpy as np # module for low-level scientific computing.
import scipy as sp # library for working with NumPy arrays
import scipy.io as sio # read/write data to various formats
import scipy.signal # signal processing module
import pandas as pd # data manipulation tools. built on NumPy library
import os # functions for interacting w operating system
import matplotlib.pyplot as plt # makes matplotlib work like MATLAB. ’pyplot’ functions. 

##### SETTINGS #####
# file = 'joels_ECoG_data.mat' # file to load
my_file = 'emodat.mat'

####################

# read MATLAB file
my_file = os.path.join('./', file) # concatenate
my_data = sio.loadmat(my_file)
sample_rate = my_data['srate']
sample_rate = sample_rate[0][0];
time_series = my_data['data']
time_series = time_series[0,:]
event_inds = [1034, 2133, 4681, 9960, 12574] #Time indicies when a stimulus is presented to the subject

print("my_data has type ", type(my_data))
print("there are (", str(len(time_series)), ") samples")
print("sampling rate is ~(", round(sample_rate,2),') samples/second')
print('the event indices are:')
print(*event_inds, sep=',')
```

```python
##### SETTINGS #####
window = [500,1000] # [samples before event, samples after event]
event_ind = 2 # choose which event

this_event = event_inds[event_ind]
win_length = window[0] + window[1] # number of samples in window
win_time = win_length / sample_rate # window size in seconds

print('\nWe will look at the event w/ index (',event_ind,')')
print('Range will be (',window[0],') samples before and (',window[1], ') samples after')
print('...so the range of samples we want is ',this_event,'-',window[0],' : ',this_event,'+',window[1])
print('The window spans ~(',round(win_time,2),') seconds')

my_ERP = time_series[this_event-window[0]:this_event + window[1]]
plt.plot(my_ERP,label = 'ERP')
plt.plot([500,500], [-150, 100],label = 'event onset')
legend();
plt.show()
print("ECoG data ~500 ms before and ~1000 ms after an event")
```

```python
# plotting all ERPs on single plot...

for event in range(len(event_inds)):
    this_ERP = time_series[event_inds[event]-window[0]:event_inds[event] + window[1]]
    plot(this_ERP,label = 'ERP ' + str(event+1))

plot([500,500], [-200, 200],label = 'event onset')    
legend();
plt.show()
```

```python
# the above plots are not helpful at all. need to subtract baseline...

chan_baselines = zeros(len(event_inds)); # stores baseline ERP data
ERP_matrix = zeros((len(event_inds),window[0]+window[1])); # will store all trials to obtain the mean

for event in range(len(event_inds)):
    chan_baselines[event] = mean(time_series[event_inds[event]-100:event_inds[event] + 0]) 
    ERP_baseline_removed = time_series[event_inds[event]-window[0]:event_inds[event] + window[1]] - chan_baselines[event] #subtracting baseline from each trial.    
    ERP_matrix[event,:] = ERP_baseline_removed; 
    
mean_ERP = mean(ERP_matrix,0) # mean voltage across trials

print('Baselines for each event:')
print(chan_baselines)

plot(mean_ERP, label = 'ERP');
plot([500,500], [-60, 60],label = 'event onset');
legend();
plt.show()
```
