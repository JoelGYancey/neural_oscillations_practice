- sampling rate: 1 kHz
- units: microvolts (μV)

```python
# We can use the signal processing module (scipy.signal) to estimate the power spectral density (PSD).
# PSD shows how the strength of a signal is distributed across a range of frequencies.
# In this case, we are interested in the range of frequencies where neural oscillations occur.

f, psd = sp.signal.welch(lfp_data[:,0], Fs, nperseg=1000)

# Plot the power spectrum
plt.figure(figsize=(11,3))
plt.semilogy(f,psd,'k')
sns.despine()
plt.xlim((0,200))
plt.yticks(size=15)
plt.xticks(size=15)
plt.ylabel('power ($uV^{2}/Hz$)',size=15)
plt.xlabel('frequency (Hz)',size=15)
plt.title('PSD of Local Field Potential', size=20)
plt.show()
```

```python
# We will now generate and plot a time-frequency profile (spectrogram) for a range of data. 

# creates spectrogram over a range of data
samp_spec = range(10000,20000)
f, t_spec, x_spec = sp.signal.spectrogram(lfp_data[samp_spec,0], fs=Fs, window='hanning', nperseg=1000, noverlap=1000-1, mode='psd')
fmax = 200

x_mesh, y_mesh = np.meshgrid(t_spec, f[f<fmax])
plt.figure(figsize=(11,4))
plt.subplot(2,1,1)
plt.title('Time-Frequency Profile of Local Field Potential', size=20)
#plt.xlim()
plt.pcolormesh(x_mesh+t_index[samp_spec[0]], y_mesh, np.log10(x_spec[f<fmax]), cmap=cm.jet)#, vmin=vmin, vmax=vmax)
plt.ylabel('frequency (Hz)')
#plt.colorbar()

plt.subplot(2,1,2)
plt.plot(t_index[samp_spec],lfp_data[samp_spec],'k')
plt.xlabel('time (s)')
plt.ylabel('voltage (a.u.)')


plt.show()

```
